```python
def remove_punct(text):
    
    text_nopunct = ''.join([char for char in text if char not in string.punctuation])
    return text_nopunct

dataset['body_text_clean'] = dataset['body_text'].apply(lambda x: remove_punct(x))
dataset.head()
```

## TOKENIZING

```python
# import re

# def tokenize(text):
#     tokens = re.split('\W+', text)
#     return tokens

# dataset['body_text_tokenized'] = dataset['body_text_clean'].apply(lambda x: tokenize(x))
# dataset.head()
```

```python
import re

def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

dataset['body_text_tokenized'] = dataset['body_text_clean'].apply(lambda x: tokenize(x.lower()))
dataset.head()
```

```python
#REMOVING STOPWORDS
```

```python
import nltk
stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(tokenize_list):
    text = [word for word in tokenize_list if word not in stopword]
    return text

dataset['body_text_nostop'] = dataset['body_text_tokenized'].apply(lambda x: remove_stopwords(x))
dataset.tail()
```

## STEMMING

```python
ps = nltk.PorterStemmer()
```

```python
print(ps.stem('grows'))
print(ps.stem('growing'))
print(ps.stem('grow'))
```

```python
print(ps.stem('run'))
print(ps.stem('running'))
print(ps.stem('runner'))
```

```python
def stemming(stopped_text):
    
    text = [ps.stem(word) for word in stopped_text]
    return text

dataset['body_text_stemmed'] = dataset['body_text_nostop'].apply(lambda x: stemming(x))
dataset.head()
dataset.tail()
```

## Lemmatizing

```python
wn = nltk.WordNetLemmatizer()
nltk.download('wordnet')
```

```python
print(ps.stem('meanness'))
print(ps.stem('meaning'))

print(wn.lemmatize('meanness'))
print(wn.lemmatize('meaning'))
```

```python
print(ps.stem('geese'))
print(ps.stem('goose'))

print(wn.lemmatize('geese'))
print(wn.lemmatize('goose'))
```

```python
def lemmatizing(tokenized_text):
    
    text = [wn.lemmatize(word) for word in tokenized_text]
    return text

dataset['body_text_lemmatized'] = dataset['body_text_nostop'].apply(lambda x: lemmatizing(x))
dataset.head()
```

## COUNT VECTORIZING

```python
##From Starting again in a single code:

def clean_text(text):
    
    text = ''.join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [ps.stem(word) for word in tokens if word not in stopword]
    return text
```

```python
# Please note that all the steps performed are done in single step in above cell.
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer(analyzer=clean_text)
X_counts = count_vect.fit_transform(dataset['body_text'])
X_counts.shape
```

```python
print(count_vect.get_feature_names())
```

```python
data_sample = dataset[0:20]

count_vect_sample = CountVectorizer(analyzer=clean_text)
X_counts_sample = count_vect_sample.fit_transform(data_sample['body_text'])
print(X_counts_sample.shape)
print(count_vect_sample.get_feature_names())
```

## NGRAM EXPERIMENTING

```python
##From Starting again in a single code:
##Here we are adding back the splitted words to a single string

def clean_gram(text):
    
    text = ''.join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = ' '.join([ps.stem(word) for word in tokens if word not in stopword])
    return text
```

```python
dataset['cleaned_gram'] = dataset['body_text'].apply(lambda x: clean_gram(x))
```

```python
dataset.head()
```

```python
##Applying NGRAM with CV

ngram_vect = CountVectorizer(ngram_range=(2,2))
X_counts_gram = ngram_vect.fit_transform(dataset['cleaned_gram'])
print(X_counts_gram.shape)
ngram_vect.get_feature_names()
```

```python
data_sample = dataset[0:20]

count_vect_gram = CountVectorizer(ngram_range=(2,2))
X_counts_gram_small = count_vect_gram.fit_transform(data_sample['cleaned_gram'])
print(X_counts_gram_small.shape)
print(count_vect_gram.get_feature_names())
```

```python
X_counts_df_gram = pd.DataFrame(X_counts_gram_small.toarray(), columns = count_vect_gram.get_feature_names())

X_counts_df_gram
```

## TFIDF 

```python
##From Starting again in a single code:

def clean_tfidf(text):
    
    text = ''.join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [ps.stem(word) for word in tokens if word not in stopword]
    return text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect = TfidfVectorizer(analyzer=clean_tfidf)
X_tfidf = tfidf_vect.fit_transform(dataset['body_text'])
print(X_tfidf.shape)
tfidf_vect.get_feature_names()
```

```python
data_sample = dataset[0:20]

tfidf_vect_sample = TfidfVectorizer(analyzer=clean_tfidf)
X_tfidf_sample = tfidf_vect_sample.fit_transform(data_sample['body_text'])
print(X_tfidf_sample.shape)
tfidf_vect_sample.get_feature_names()
```

```python
X_tfidf_df = pd.DataFrame(X_tfidf_sample.toarray(), columns=tfidf_vect_sample.get_feature_names())
X_tfidf_df
```
